# Working with grades data

  a) Draw a line chart for proportion of students that are missing approved grade for one or more subjects for the years 18-23.

```python
import pandas as pd

data_file = "data/betyg_o_prov_riksnivå.xlsx"

df = pd.read_excel(data_file, sheet_name= "Tabell 1B", skiprows=7)

df
```

```python
df_rename = df.rename(
    {
        "Unnamed: 0": "Huvudman",
        "Totalt": "Totalt-16 ämne",
        "Flickor": "Flickor-16 ämne",
        "Pojkar": "Pojkar-16 ämne",
        "Totalt.1": "Totalt-17 ämne",
        "Flickor.1": "Flickor-17 ämne",
        "Pojkar.1": "Pojkar-17 ämne",
        "Totalt.2": "Totalt-saknar godkänt",
        "Flickor.2": "Flickor-saknar godkänt",
        "Pojkar.2": "Pojkar-saknar godkänt",
    },
    axis=1,
)
```

```python
df_rename
```

```python
year_total = df_rename.head()
year_total.head()
```

```python
year_total_list = list(year_total["Totalt-saknar godkänt"])
```

```python
year = list(df_rename["Huvudman"].head())
year
```

```python
import matplotlib.pyplot as plt

plt.plot(year, year_total_list, label= "difference in year")
plt.xlabel("Year")
plt.ylabel("percentage missing approved")
plt.title("How many have not passing")
plt.legend()
```

```python
data_not_passing =year_total[["Huvudman", "Totalt-saknar godkänt"]]
data_not_passing
```

```python
import seaborn as sns

sns.barplot(data=data_not_passing, x="Huvudman", y="Totalt-saknar godkänt")
```

```python
sns.lineplot(data=data_not_passing, x="Huvudman", y= "Totalt-saknar godkänt", label= "How it change per year")
plt.title("Percenatge failer")
plt.legend()
```
