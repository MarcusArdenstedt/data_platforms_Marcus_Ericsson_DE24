## Data cleaning

```python
df.columns
```

```python
df.rename(
    {"Cals_per100grams": "calories", "KJ_per1oograms": "KJ", "per100grams": "per100"},
    axis=1,
)
```

```python
df_cleaned_name = df.rename(
    {"Cals_per100grams": "calories", "KJ_per1oograms": "KJ", "per100grams": "per100"},
    axis=1,
)
df_cleaned_name.head()
```

```python
df_cleaned_name["calories"] = df_cleaned_name["calories"].str[:-3].astype(int)
```

```python
df_cleaned_name.head()
```

```python
df_cleaned_name.info()
```

```python
df_cleaned_name["per100"].value_counts()
```

```python
df_cleaned_name["per100"] == "100ml"
```

```python
df_liquids = df_cleaned_name[df_cleaned_name["per100"] == "100ml"]
```

```python
df_liquids.head()
```

```python
df_solids = df_cleaned_name.query("per100 == '100g'")
df_solids.head()
```

## Find top 5 categories of highest calories

```python
df_solids_sorted = df_solids.sort_values(by="calories", ascending=False)
df_solids_top5 = df_solids_sorted.iloc[:5]
df_solids_top5
```

```python
df_liquids_top5 = df_liquids.sort_values(by="calories", ascending=False).head()
df_liquids_top5
```

```python
df_top5_categories = df_cleaned_name.groupby("FoodCategory")["calories"].median().sort_values(
    ascending=False
).reset_index().head()

df_top5_categories
```

```python
query = """
SELECT
    FoodCategory,
    Median(calories) as median_calories
From df_cleaned_name
GROUP BY 
    FoodCategory
ORDER BY 
    median_calories
DESC
   limit 5
"""
```

```python
import duckdb
duckdb.query(query).df()
```
